```python
import numpy as np
import spectrally_constrained_LVMs as sc_LVMs
from matplotlib import pyplot as plt

# Load in the time series signal
Fs = 20480
#)

data_dict = np.loadtxt("./2004.02.16.04.32.39")
x_signal = data_dict[:, 0]

# Perform pre-processing
Lw = 256
Lsft = 1
X = sc_LVMs.Hankel_matrix(x_signal, Lw, Lsft) # Hankelise the vibration data

# Define the cost function
cost_inst = sc_LVMs.negentropy_cost("exp", {"alpha":1}) # ICA objective

# Define the model
model_inst = sc_LVMs.linear_model(n_sources = 10, 
                                  cost_instance=cost_inst, 
                                  whiten = True, 
                                  organise_by_kurtosis=True,
                                  alpha_reg=10,
                                  use_ls = True, 
                                  verbose = True)

# Estimate the model parameters
model_inst.fit(X, n_iters = 500, learning_rate = 1, tol = 1e-4, Fs = Fs)
```
